## 🎯 O Problema

Pegar 4 pseudocódigos escritos em sala (lógica em português estruturado) e traduzir cada um para Python funcional, validando as estruturas de **entrada**, **decisão**, **repetição** e **acumulação**.

## 🧩 Os 4 Algoritmos

| # | Função | Conceito treinado |
|---|---|---|
| 1 | `processar_vendas()` | Repetição com `for`, acumulador, condicional `if/else` |
| 2 | `analisar_clima()` | Decisão em cascata (`if/elif/else`) com faixas de valor |
| 3 | `sistema_notas_turma()` | Loop aninhado (turma × provas) + média + classificação |
| 4 | `simulador_poupanca()` | Loop com cálculo composto (juros mensais acumulados) |

## 🛠️ Função 1 — Processar Vendas

Lê N vendas, soma o total e classifica cada uma como ALTA (≥ 1000) ou BAIXA.

```python
def processar_vendas():
    quantidade = int(input("Quantas vendas deseja processar? "))
    total = 0

    for i in range(1, quantidade + 1):
        valor = float(input(f"Digite o valor da venda {i}: R$ "))
        total = total + valor

        if valor >= 1000:
            print(f"Venda {i}: ALTA")
        else:
            print(f"Venda {i}: BAIXA")

    print(f"\nTotal vendido: R$ {total}")

processar_vendas()
```

## 🌡️ Função 2 — Analisar Clima

Recebe a temperatura e classifica em FRIO (< 15), AMENO (15–24), QUENTE (25–34) ou MUITO QUENTE (≥ 35).

```python
def analisar_clima():
    temperatura = float(input("Digite a temperatura em °C: "))

    if temperatura < 15:
        print("FRIO")
    elif temperatura < 25:
        print("AMENO")
    elif temperatura < 35:
        print("QUENTE")
    else:
        print("MUITO QUENTE")

analisar_clima()
```

## 🎓 Função 3 — Sistema de Notas da Turma

Lê N alunos × 3 provas cada, calcula a média individual e classifica como APROVADO (≥ 7), RECUPERAÇÃO (5–6,9) ou REPROVADO.

```python
def sistema_notas_turma():
    quantidade_alunos = int(input("Quantos alunos têm na turma? "))

    for aluno in range(1, quantidade_alunos + 1):
        print(f"\n--- Aluno {aluno} ---")
        soma_notas = 0

        for prova in range(1, 4):
            nota = float(input(f"Nota da prova {prova}: "))
            soma_notas = soma_notas + nota

        media = soma_notas / 3
        print(f"Média: {media}")

        if media >= 7:
            print("APROVADO")
        elif media >= 5:
            print("RECUPERAÇÃO")
        else:
            print("REPROVADO")

sistema_notas_turma()
```

## 💰 Função 4 — Simulador de Poupança

Simula juros compostos mensais a partir de um valor inicial, taxa mensal e prazo em meses.

```python
def simulador_poupanca():
    valor_inicial = float(input("Valor inicial: R$ "))
    taxa_mensal = float(input("Taxa de juros mensal (%): "))
    meses = int(input("Quantos meses? "))

    saldo = valor_inicial
    taxa_decimal = taxa_mensal / 100

    for mes in range(1, meses + 1):
        rendimento = saldo * taxa_decimal
        saldo = saldo + rendimento
        print(f"Mês {mes}: R$ {saldo}")

    lucro_total = saldo - valor_inicial
    print(f"\nValor final: R$ {saldo}")
    print(f"Lucro total: R$ {lucro_total}")

simulador_poupanca()
```

### Pergunta 1 — Sobre tipagem dinâmica e `input()`

**Por que precisamos converter o retorno de `input()` com `float()` ou `int()` antes de fazer contas?**

> O `input()` sempre recebe o que você digita como texto, mesmo que seja um número. Se esquecer o `float()` ou `int()`, o Python não consegue fazer conta com texto e dá erro. Por isso tem que converter sempre que for usar o número em cálculo.

### Pergunta 2 — Sobre `range()` e o limite superior

**Qual a diferença entre `DE 1 ATE 5` no pseudocódigo e `range(1, 5)` no Python?**

> No pseudocódigo quando escrevemos `DE 1 ATE 5` o 5 entra na conta. No Python o `range(1, 5)` para no 4 e não inclui o 5. Por isso sempre colocamos `range(1, quantidade + 1)` pra não perder o último número da contagem.
