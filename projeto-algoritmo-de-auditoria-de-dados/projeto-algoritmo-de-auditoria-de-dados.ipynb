```python
# Variável Global
LIMITE_SEGURANCA = 10000.00

print(LIMITE_SEGURANCA)
```

```python
# 3 Vendas do dia ao Usuário

venda1 = float(input("Digite o valor da venda 1: R$ "))
venda2 = float(input("Digite o valor da venda 2: R$ "))
venda3 = float(input("Digite o valor da venda 3: R$ "))
```

```python
# Analisar Vendas do Dia
def analisar_vendas(venda1, venda2, venda3):

# Alterar Variável Global Dentro da Função
  global LIMITE_SEGURANCA

# Calculo da Média
soma = venda1 + venda2 + venda3
media = soma / 3

# Valores e Tipos
print("RELATÓRIO DE VENDAS")
print(f"Venda 1: R$ {venda1} | tipo: {type(venda1)}")
print(f"Venda 2: R$ {venda2} | tipo: {type(venda2)}")
print(f"Venda 3: R$ {venda3} | tipo: {type(venda3)}")
print(f"Média do dia: R$ {media}")

# Verificando se média passou do limite
if media > LIMITE_SEGURANCA:
  print("ALERTA: SISTEMA EM QUARENTENA")

# Verificando se a Venda é 5x Maior que a Média
if venda1 > media * 5:
  print("Venda 1: REVISÃO MANUAL")
if venda2 > media * 5:
  print("Venda 2: REVISÃO MANUAL")
if venda3 > media * 5:
  print("Venda 3: REVISÃO MANUAL")

# Chamando a Função com valores digitados
analisar_vendas(venda1, venda2, venda3)




```
