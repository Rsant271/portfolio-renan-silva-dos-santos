## 🟠 Desafio 1 — O Criador de Emojis (Pixels RGB)

**Tema escolhido:** logo do Claude Code em pixel art 5×5.

- **Fundo:** branco `(255, 255, 255)`
- **Asterisco/raios:** laranja Claude `(217, 119, 87)`

Aplicação de filtro de sombra: cada pixel laranja tem o brilho reduzido em 50% (divisão inteira por 2). Como tuplas são imutáveis, é montada uma nova tupla para cada pixel alterado.

```python
import matplotlib.pyplot as plt

# Estrutura base: emoji do Claude Code em 5x5
# Fundo branco (255,255,255), asterisco laranja Claude (217,119,87)
emoji_data = {
    "nome": "Claude Code",
    "grade": [
        [(255,255,255), (255,255,255), (217,119,87),  (255,255,255), (255,255,255)],  # Linha 0
        [(217,119,87),  (255,255,255), (217,119,87),  (255,255,255), (217,119,87) ],  # Linha 1 (Raios)
        [(255,255,255), (217,119,87),  (217,119,87),  (217,119,87),  (255,255,255)],  # Linha 2 (Centro)
        [(217,119,87),  (255,255,255), (217,119,87),  (255,255,255), (217,119,87) ],  # Linha 3 (Raios)
        [(255,255,255), (255,255,255), (217,119,87),  (255,255,255), (255,255,255)]   # Linha 4
    ]
}

print("Emoji original:")
plt.imshow(emoji_data["grade"])
plt.show()

nova_grade = []

# Nivel 1: Percorrendo o Dicionario (Lote)
for chave, valor in emoji_data.items():
    if chave == "grade":
        # Nivel 2: Percorrendo as Linhas
        for linha in valor:
            nova_linha = []
            # Nivel 3: Percorrendo os Pixels
            for pixel in linha:
                # Logica do Filtro: Se for Laranja Claude (217, 119, 87)
                if pixel == (217, 119, 87):
                    # Criando nova tupla com brilho reduzido (50%)
                    pixel_sombreado = (pixel[0] // 2, pixel[1] // 2, pixel[2] // 2)
                    nova_linha.append(pixel_sombreado)
                else:
                    # Se for branco, mantem original
                    nova_linha.append(pixel)

            nova_grade.append(nova_linha)

emoji_data["grade"] = nova_grade

print(f"Emoji: {emoji_data['nome']} (com filtro de sombra)")
for linha in emoji_data["grade"]:
    print(linha)

plt.imshow(emoji_data["grade"])
plt.show()
```

## 🎵 Desafio 2 — Matrizes Musicais (Transposição)

**Tema escolhido:** tons das notificações sonoras do Claude Code (Start, Success, Error). Cada notificação guarda uma matriz 2×2 de frequências em Hz.

**Operação:** transposição (linha vira coluna). Como tuplas são imutáveis, monta-se uma nova tupla com as listas invertidas e usa-se `dict.update()` para atualizar o valor da chave.

```python
# Desafio 2 - Transposicao das matrizes de tons do Claude Code

# Entrada - biblioteca com os tons das notificacoes sonoras do Claude Code
biblioteca_sonora = {
    "Notificacoes": [
        {"Start":   ([440, 480], [520, 560])},  # tom de inicio de sessao
        {"Success": ([600, 660], [720, 780])},  # tom de task concluida
        {"Error":   ([200, 150], [100, 50] )}   # tom de erro
    ]
}

# Processamento - transpoe cada matriz 2x2 (linha vira coluna)
# Nivel 1: percorre a lista de notificacoes
for notificacao in biblioteca_sonora["Notificacoes"]:
    # Nivel 2: percorre o dicionario da notificacao
    for nome, matriz in notificacao.items():
        nova_linha_0 = []
        nova_linha_1 = []
        # Nivel 3: percorre cada linha da matriz original
        for linha in matriz:
            # Primeiro elemento da linha vai para a nova linha 0
            nova_linha_0.append(linha[0])
            # Segundo elemento da linha vai para a nova linha 1
            nova_linha_1.append(linha[1])
        # Tupla e imutavel, entao monto uma nova com as listas invertidas
        nova_matriz = (nova_linha_0, nova_linha_1)
        # .update() troca o valor da chave dentro do dicionario
        notificacao.update({nome: nova_matriz})

# Saida
print("Matrizes de tons apos a transposicao:")
print("-" * 40)
for notificacao in biblioteca_sonora["Notificacoes"]:
    for nome, matriz in notificacao.items():
        print(f"{nome}: {matriz}")
```

## 🤖 Desafio 3 — O Integrador (Galeria do Claude)

**Tema escolhido:** galeria dos modelos disponíveis do Claude (Opus 4.7, Sonnet 4.6, Haiku 4.5).

**Estrutura:** dicionário aninhado em lista, com tuplas para nome e tags, e lista de tuplas para benchmarks. 

**Métodos usados:** `.keys()` (dict), `.count()` (tupla), `.pop()` (lista), `.insert()` (lista), `.update()` (dict).

```python
# Desafio 3 - Galeria de Modelos do Claude

# Entrada - biblioteca com os modelos disponiveis do Claude
# Cada modelo tem nome (tupla), benchmarks (lista de tuplas) e tags (tupla)
galeria_claude = {
    "Modelos_Disponiveis": [
        {
            "nome": ("Opus", "4.7"),
            "benchmarks": [(95, 92, 88), (90, 85, 80)],
            "tags": ("raciocinio", "premium", "raciocinio", "agente")
        },
        {
            "nome": ("Sonnet", "4.6"),
            "benchmarks": [(88, 90, 85), (87, 88, 82)],
            "tags": ("rapido", "balanceado", "rapido", "rapido")
        },
        {
            "nome": ("Haiku", "4.5"),
            "benchmarks": [(75, 80, 70), (72, 78, 68)],
            "tags": ("rapido", "economico", "rapido")
        }
    ]
}

# Processamento - monta um relatorio de performance dos modelos
relatorio = []

# Nivel 1: percorre as chaves do dicionario principal com .keys()
for nome_grupo in galeria_claude.keys():
    # Nivel 2: percorre a lista de modelos
    for modelo in galeria_claude[nome_grupo]:
        # Soma os scores de codigo dos benchmarks
        total_codigo = 0
        # Nivel 3: percorre cada linha de benchmark (tupla)
        for linha in modelo["benchmarks"]:
            # O segundo valor da tupla e o score de codigo
            total_codigo += linha[1]
        # .count() da tupla de tags conta quantas vezes a primeira tag aparece
        primeira_tag = modelo["tags"][0]
        vezes = modelo["tags"].count(primeira_tag)
        # Monta a linha do relatorio
        linha_rel = f"{modelo['nome'][0]} {modelo['nome'][1]} - score codigo: {total_codigo} | tag '{primeira_tag}' x{vezes}"
        relatorio.append(linha_rel)

# .pop() remove o ultimo modelo do relatorio
ultimo_removido = relatorio.pop()
# .insert() coloca o cabecalho no topo da lista
relatorio.insert(0, "=== RELATORIO DA GALERIA CLAUDE ===")
# .update() adiciona um campo novo no dicionario principal
galeria_claude.update({"total_modelos": 3})

# Saida
print("-" * 40)
for linha in relatorio:
    print(linha)
print("-" * 40)
print(f"Modelo removido do relatorio: {ultimo_removido}")
print(f"Total de modelos na galeria: {galeria_claude['total_modelos']}")
```
