## 🎯 O Problema

Escrever um algoritmo que lê um valor de compra e o valor pago pelo cliente, calcula o troco total e determina o **menor número possível de notas** (R$ 100, R$ 50, R$ 10, R$ 5 e R$ 1) para devolver ao cliente.

## 🧩 A Estratégia: Modularização

Em vez de escrever um único bloco gigante de código ("código espaguete"), o problema foi dividido em três funções independentes, cada uma com uma única responsabilidade (alta coesão e baixo acoplamento):

| Função | Responsabilidade |
|---|---|
| `validar_pagamento(compra, pago)` | Verifica se o valor pago é suficiente. |
| `calcular_troco(compra, pago)` | Retorna a diferença matemática bruta. |
| `decompor_notas(troco)` | Processa a divisão e os restos para cada denominação de nota. |

## 🗺️ Fluxograma da Solução

```mermaid
flowchart TD
    A([INÍCIO]) --> B[/Ler valor_compra/]
    B --> C[/Ler valor_pago/]
    C --> D{validar_pagamento<br/>compra, pago?}
    D -->|FALSO| E[/Mostrar: Pagamento insuficiente/]
    E --> Z([FIM])
    D -->|VERDADEIRO| F[troco ← calcular_troco<br/>compra, pago]
    F --> G[notas ← decompor_notas<br/>troco]
    G --> H[/Mostrar Total, Pago, Troco/]
    H --> I[/Mostrar quantidade de cada nota/]
    I --> Z
```

**Leitura do fluxograma:**
1. O programa começa lendo dois valores: o total da compra e o valor pago.
2. Antes de calcular o troco, valida se o pagamento é suficiente. Se não for, encerra com uma mensagem.
3. Sendo válido, calcula o troco e decompõe em notas usando algoritmo guloso (sempre tenta a maior nota primeiro).
4. Mostra o relatório final ao caixa.

## 📜 Pseudocódigo (Top-Down)

### Programa Principal (Integrador)

```
PROGRAMA SistemaCaixaTroco:
    INÍCIO
        ESCREVER "=== SISTEMA DE CAIXA ==="
        valor_compra ← LER("Total da compra: R$ ")
        valor_pago   ← LER("Valor pago:      R$ ")

        SE validar_pagamento(valor_compra, valor_pago) ENTÃO
            troco ← calcular_troco(valor_compra, valor_pago)
            notas ← decompor_notas(troco)

            ESCREVER "TOTAL: R$", valor_compra
            ESCREVER "PAGO:  R$", valor_pago
            ESCREVER "TROCO: R$", troco
            ESCREVER "---"
            PARA CADA (denominacao, quantidade) EM notas FAÇA
                SE quantidade > 0 ENTÃO
                    ESCREVER quantidade, "x R$", denominacao
                FIM-SE
            FIM-PARA
        SENÃO
            ESCREVER "Pagamento insuficiente. Operação cancelada."
        FIM-SE
    FIM
FIM-PROGRAMA
```

### Funções Subordinadas (Especialista)

```
FUNÇÃO validar_pagamento(compra, pago) → BOOLEANO:
    INÍCIO
        SE pago >= compra ENTÃO
            RETORNAR VERDADEIRO
        SENÃO
            RETORNAR FALSO
        FIM-SE
    FIM
FIM-FUNÇÃO

FUNÇÃO calcular_troco(compra, pago) → NÚMERO:
    INÍCIO
        diferenca ← pago - compra
        RETORNAR diferenca
    FIM
FIM-FUNÇÃO

FUNÇÃO decompor_notas(troco) → DICIONÁRIO:
    INÍCIO
        denominacoes ← [100, 50, 10, 5, 1]
        notas ← DICIONÁRIO_VAZIO
        valor_restante ← troco

        PARA CADA d EM denominacoes FAÇA
            quantidade ← INTEIRO(valor_restante / d)
            notas[d] ← quantidade
            valor_restante ← valor_restante - (quantidade × d)
        FIM-PARA

        RETORNAR notas
    FIM
FIM-FUNÇÃO
```

### ✅ Como esse pseudocódigo respeita os 4 critérios da Matriz de Avaliação

1. **DRY (Don't Repeat Yourself):** a matemática repetitiva das notas foi encapsulada em um único `PARA` dentro de `decompor_notas()`, não há código duplicado por denominação.
2. **Coesão Funcional:** cada função faz exatamente uma operação (validar / calcular / decompor).
3. **Isolamento de Escopo:** as funções recebem dados por parâmetros e devolvem por retorno. Não há variáveis globais poluindo o sistema.
4. **Clareza Top-Down:** o programa principal conta a história do fluxo (ler → validar → calcular → decompor → mostrar) sem se perder em detalhes matemáticos.

## 🐍 Implementação em Python (Bônus)

O pseudocódigo acima é o entregável principal. As células a seguir traduzem o mesmo desenho para Python como exercício de validação.

````python
# Função 1 — Valida se o pagamento é suficiente
def validar_pagamento(compra, pago):
    return pago >= compra
````

````python
# Função 2 — Calcula o troco bruto
def calcular_troco(compra, pago):
    return pago - compra
````

````python
# Função 3 — Decompõe o troco em notas (algoritmo guloso)
def decompor_notas(troco):
    denominacoes = [100, 50, 10, 5, 1]
    notas = {}
    valor_restante = troco

    for d in denominacoes:
        quantidade = int(valor_restante // d)
        notas[d] = quantidade
        valor_restante = valor_restante - (quantidade * d)

    return notas
````

````python
# Programa Principal — orquestra as 3 funções
print("=== SISTEMA DE CAIXA ===")
valor_compra = float(input("Total da compra: R$ "))
valor_pago   = float(input("Valor pago:      R$ "))

if validar_pagamento(valor_compra, valor_pago):
    troco = calcular_troco(valor_compra, valor_pago)
    notas = decompor_notas(troco)

    print(f"TOTAL: R$ {valor_compra:.2f}")
    print(f"PAGO:  R$ {valor_pago:.2f}")
    print(f"TROCO: R$ {troco:.2f}")
    print("---")
    for denominacao, quantidade in notas.items():
        if quantidade > 0:
            print(f"{quantidade}x R$ {denominacao}")
else:
    print("Pagamento insuficiente. Operação cancelada.")
````
