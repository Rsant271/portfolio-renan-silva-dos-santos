## 🌡️ Parte 1 — O Algoritmo do Microclima Local

**Pesquisa física:** três pontos no trajeto Pq. Novo Mundo → Sé → Tatuapé (UNICID), em dois horários diferentes do dia 17/04/2026 (08h e 18h).

**Fontes:** CETESB QUALAR (IQA) + CGE-SP / Climatempo (temperatura e umidade).

**Faixas oficiais de IQA (CETESB / CONAMA 491/2018):**

| Faixa | Limite | Risco à saúde |
|---|---|---|
| Boa | 0–40 | Sem riscos |
| Moderada | 41–80 | Pessoas sensíveis podem ter sintomas leves |
| Ruim | 81–120 | Toda a população pode apresentar sintomas |
| Muito Ruim | 121–200 | Efeitos adversos ao coração e pulmão |
| Péssima | >200 | Sérios riscos à saúde |

```python
# Constantes das faixas oficiais de IQA no Brasil (CETESB / CONAMA 491/2018)
# Cada constante representa o LIMITE SUPERIOR daquela faixa
LIMITE_BOA = 40
LIMITE_MODERADA = 80
LIMITE_RUIM = 120
LIMITE_MUITO_RUIM = 200

# Pesos da formula da Nota de Conforto Urbano
PESO_TEMPERATURA = 0.30
PESO_UMIDADE = 0.05
PESO_IQA = 0.04

# Valores considerados ideais (centro do conforto)
TEMPERATURA_IDEAL = 24
UMIDADE_IDEAL = 50


def classificar_iqa(iqa):
    # Transforma o valor numerico do IQA em uma faixa (1 a 5)
    if iqa <= LIMITE_BOA:
        return 1
    if iqa <= LIMITE_MODERADA:
        return 2
    if iqa <= LIMITE_RUIM:
        return 3
    if iqa <= LIMITE_MUITO_RUIM:
        return 4
    return 5


def descrever_iqa(faixa):
    # match-case: retorna a descricao oficial da faixa
    # O caractere | significa 'ou'
    # O _ e o caso default
    match faixa:
        case 1:
            return "Boa - sem riscos a saude"
        case 2:
            return "Moderada - pessoas sensiveis podem ter sintomas leves"
        case 3:
            return "Ruim - toda a populacao pode apresentar sintomas"
        case 4:
            return "Muito Ruim - efeitos adversos ao coracao e pulmao"
        case 5 | _:
            return "Pessima - serios riscos a saude"


def calcular_nota_conforto(temperatura, umidade, iqa):
    # Formula propria: nota comeca em 10 e cada fator gera penalidade

    # Penalidade da temperatura: quanto mais longe de 24 C, pior
    if temperatura > TEMPERATURA_IDEAL:
        penalidade_temp = (temperatura - TEMPERATURA_IDEAL) * PESO_TEMPERATURA
    else:
        penalidade_temp = (TEMPERATURA_IDEAL - temperatura) * PESO_TEMPERATURA

    # Penalidade da umidade: quanto mais longe de 50%, pior
    if umidade > UMIDADE_IDEAL:
        penalidade_umid = (umidade - UMIDADE_IDEAL) * PESO_UMIDADE
    else:
        penalidade_umid = (UMIDADE_IDEAL - umidade) * PESO_UMIDADE

    # Penalidade do IQA
    penalidade_iqa = iqa * PESO_IQA

    nota = 10 - penalidade_temp - penalidade_umid - penalidade_iqa

    # Operadores relacionais: limitar a nota entre 0 e 10
    if nota < 0:
        nota = 0
    elif nota > 10:
        nota = 10

    return nota


def analisar_microclima():
    print("=" * 60)
    print("ANALISE DE MICROCLIMA LOCAL")
    print("Trajeto: Pq. Novo Mundo -> Se -> Tatuape (UNICID)")
    print("Coleta em 17/04/2026 as 08h e 18h")
    print("=" * 60)

    # Lista de listas com os dados coletados em campo
    # Fontes: CETESB QUALAR (IQA) + CGE-SP / Climatempo (temperatura e umidade)
    # [nome_do_local, temperatura, umidade, iqa, horario]
    dados_locais = [
        ["Av. Tenente Amaro F. da Silveira - Pq. Novo Mundo", 17, 85, 60, "08h"],
        ["Av. Tenente Amaro F. da Silveira - Pq. Novo Mundo", 23, 62, 78, "18h"],
        ["Praca da Se - Centro",                              18, 75, 72, "08h"],
        ["Praca da Se - Centro",                              24, 55, 92, "18h"],
        ["Praca Silvio Romero - Tatuape (proximo a UNICID)", 17, 82, 48, "08h"],
        ["Praca Silvio Romero - Tatuape (proximo a UNICID)", 22, 68, 55, "18h"],
    ]

    # Rotulos das metricas numericas (usados no for aninhado)
    metricas = ["Temperatura (C)", "Umidade (%)", "IQA"]

    # FOR ANINHADO
    # Laco externo: percorre cada local
    for local in dados_locais:
        nome = local[0]
        horario = local[4]

        print("-" * 60)
        print(f"Local: {nome}")
        print(f"Horario: {horario}")

        # Laco interno: percorre as 3 metricas numericas (posicoes 1, 2 e 3)
        for i in range(len(metricas)):
            valor = local[i + 1]
            print(f"  {metricas[i]}: {valor}")

        # Classificacao de IQA (match-case)
        iqa = local[3]
        faixa = classificar_iqa(iqa)
        descricao = descrever_iqa(faixa)
        print(f"  Qualidade do Ar: {descricao}")

        # Nota de Conforto Urbano
        temperatura = local[1]
        umidade = local[2]
        nota = calcular_nota_conforto(temperatura, umidade, iqa)
        print(f"  Nota de Conforto Urbano: {nota:.2f} / 10.00")

        # Operador logico AND: alerta duplo de calor + ar ruim
        if temperatura >= 30 and iqa > LIMITE_MODERADA:
            print(f"  [ALERTA] Calor alto E ar ruim. Evitar exposicao prolongada.")

    print("=" * 60)


analisar_microclima()
```

## 🚪 Parte 2 — Navegação e Evacuação Espacial

**Mapeamento físico:** caminhei pela minha casa e contei os nós (cômodos/corredores) entre o ponto mais distante (quarto) e a saída principal (portão da rua). Identifiquei o estado real de cada nó.

**Trajeto mapeado:** `Quarto → Sala → Cozinha → Corredor → Portão (Rua)`

**Estados reais:** a chave de casa fica no quarto e a porta da cozinha é a única trancada do trajeto — ou seja, o agente precisa pegar a chave **antes** de passar pela cozinha, senão tem que retroceder.

```python
# Constantes de controle
ENERGIA_INICIAL = 15
CUSTO_POR_MOVIMENTO = 1


def simular_evacuacao():
    print("=" * 60)
    print("SIMULADOR DE EVACUACAO - ROTA DE FUGA")
    print("=" * 60)

    # Mapa real da residencia do aluno (trajeto de fuga em emergencia)
    # Lista 1: nomes dos locais na ordem do trajeto
    locais = [
        "Quarto",
        "Sala",
        "Cozinha",
        "Corredor",
        "Portao (Rua)"
    ]

    # Lista 2: estado correspondente a cada local
    # Observacoes reais: chave de casa guardada no quarto;
    # a porta da cozinha e a unica trancada no trajeto
    estados = [
        "chave",
        "caminho livre",
        "porta trancada",
        "caminho livre",
        "saida"
    ]

    # Variaveis de controle do agente
    posicao = 0
    energia = ENERGIA_INICIAL
    inventario = []

    print(f"Energia inicial: {energia}")
    print(f"Locais mapeados: {len(locais)}")
    print("-" * 60)

    # WHILE: o agente tenta chegar na saida enquanto tem energia
    while energia > 0:
        local_atual = locais[posicao]
        estado_atual = estados[posicao]

        print(f"[Energia: {energia}] Posicao {posicao}: {local_atual} | Estado: {estado_atual}")

        # Avaliacao do estado atual
        if estado_atual == "saida":
            print("  >> Agente escapou com sucesso! <<")
            break

        elif estado_atual == "chave":
            # IF ANINHADO: ja tem a chave no inventario?
            if "chave" in inventario:
                print("  Chave ja estava no inventario. Seguindo.")
            else:
                inventario.append("chave")
                print("  Chave encontrada e guardada.")
            posicao = posicao + 1

        elif estado_atual == "caminho livre":
            print("  Caminho livre. Avancando.")
            posicao = posicao + 1

        elif estado_atual == "porta trancada":
            # IF ANINHADO: a decisao de avancar ou retroceder depende do inventario
            if "chave" in inventario:
                print("  Porta destrancada com a chave. Avancando.")
                posicao = posicao + 1
            else:
                print("  Sem chave. Retrocedendo para buscar.")
                posicao = posicao - 1

        else:
            print("  Obstaculo desconhecido. Avancando com cuidado.")
            posicao = posicao + 1

        # Cada movimento consome energia
        energia = energia - CUSTO_POR_MOVIMENTO

        # Protecao: impede sair do inicio da lista
        if posicao < 0:
            posicao = 0

    print("-" * 60)
    if energia <= 0:
        print("  >> Agente ficou sem energia antes de chegar a saida. <<")
    print(f"Energia restante: {energia}")
    print(f"Inventario final: {inventario}")
    print("=" * 60)


simular_evacuacao()
```
